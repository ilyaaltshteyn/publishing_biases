```python
import pandas as pd
```

```python
firsts = pd.read_csv('female_first_auth_by_yr_mo_counts_3plus_authors.csv').set_index('_id', drop=False)
seconds = pd.read_csv('female_second_auth_by_yr_mo_counts_3plus_authors.csv').set_index('_id')
```

```python
firsts.head()
```

```python
seconds.head()
```

```python
seconds.join(firsts, lsuffix='_seconds', rsuffix='_firsts')[2:-90]
```

```python
joined = seconds.join(firsts, lsuffix='_seconds', rsuffix='_firsts')[2:-90]
joined = joined[joined.apply(lambda x: 'undefined' not in x._id, axis=1)]

joined['abs_diff'] = joined['value_seconds'] - joined['value_firsts']
joined['percent_diff'] = joined['abs_diff']/joined['value_firsts']
joined = joined[joined.percent_diff.notnull()]
```
