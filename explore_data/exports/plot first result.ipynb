```python
import pandas as pd
```

```python
firsts = pd.read_csv('female_first_auth_by_yr_mo_counts_3plus_authors.csv').set_index('_id', drop=False)
seconds = pd.read_csv('female_second_auth_by_yr_mo_counts_3plus_authors.csv').set_index('_id')

firsts_male = pd.read_csv('male_0th_auth_papers_with_3or_more_authors_by_yr_mo_counts.csv').set_index('_id', drop=False)
seconds_male = pd.read_csv('male_1th_auth_papers_with_3or_more_authors_by_yr_mo_counts.csv').set_index('_id')
```

```python
firsts_male.head()
```

```python
seconds_male.head()
```

```python
# Female 
firsts['year'] = pd.Series(firsts.index).apply(lambda x: x.split('_')[0]).values
seconds['year'] = pd.Series(seconds.index).apply(lambda x: x.split('_')[0]).values

firsts_grouped = firsts.groupby('year').sum()
seconds_grouped = seconds.groupby('year').sum()

# Male
firsts_male['year'] = pd.Series(firsts_male.index).apply(lambda x: x.split('_')[0]).values
seconds_male['year'] = pd.Series(seconds_male.index).apply(lambda x: x.split('_')[0]).values

firsts_male_grouped = firsts_male.groupby('year').sum()
seconds_male_grouped = seconds_male.groupby('year').sum()
```

```python
# Female
joined = seconds.join(firsts, lsuffix='_seconds', rsuffix='_firsts')[2:-90]
joined = joined[joined.apply(lambda x: 'undefined' not in x._id, axis=1)]

joined_grouped = seconds_grouped.join(firsts_grouped, lsuffix='_seconds', rsuffix='_firsts')

# joined['abs_diff'] = joined['value_seconds'] - joined['value_firsts']
# joined['percent_diff'] = joined['abs_diff']/joined['value_firsts']
# joined = joined[joined.percent_diff.notnull()]

joined['total_pubs'] = joined['value_seconds'] + joined['value_firsts']
joined['probability_first'] = joined['value_firsts']/joined['total_pubs']
joined = joined[joined.probability_first.notnull()]


joined_grouped['total_pubs'] = joined_grouped['value_seconds'] + joined_grouped['value_firsts']
joined_grouped['probability_first'] = joined_grouped['value_firsts']/joined_grouped['total_pubs']
joined_grouped = joined_grouped[joined_grouped.probability_first.notnull()]
```

```python
# Male
joined_male = seconds_male.join(firsts_male, lsuffix='_seconds', rsuffix='_firsts')[2:-90]
joined_male = joined_male[joined_male.apply(lambda x: 'undefined' not in x._id, axis=1)]

joined_male_grouped = seconds_male_grouped.join(firsts_male_grouped, lsuffix='_seconds', rsuffix='_firsts')

# joined['abs_diff'] = joined['value_seconds'] - joined['value_firsts']
# joined['percent_diff'] = joined['abs_diff']/joined['value_firsts']
# joined = joined[joined.percent_diff.notnull()]

joined_male['total_pubs'] = joined_male['value_seconds'] + joined_male['value_firsts']
joined_male['probability_first'] = joined_male['value_firsts']/joined_male['total_pubs']
joined_male = joined_male[joined_male.probability_first.notnull()]


joined_male_grouped['total_pubs'] = joined_male_grouped['value_seconds'] + joined_male_grouped['value_firsts']
joined_male_grouped['probability_first'] = joined_male_grouped['value_firsts']/joined_male_grouped['total_pubs']
joined_male_grouped = joined_male_grouped[joined_male_grouped.probability_first.notnull()]
```
