```python
from rdflib import Graph
```

```python
g = Graph()
g.parse('/Users/ilya/Downloads/wikitest.ttl', format='turtle')
```

```python
cnn_articles = {}

counter = 0
for url, dbtype_long, value in g.items():
    counter += 1
    if counter % 10000 == 0:
        print 'finished', counter
    
    if url not in cnn_articles:
        cnn_articles[url] = {}
    dbtype = dbtype_long.split('/')[-1]
    if dbtype == 'title':
        cnn_articles[url]['title'] = value
    elif dbtype == 'author':
        if 'authors' not in cnn_articles[url]:
            cnn_articles[url]['authors'] = []
        cnn_articles[url]['authors'].append(value)
```
